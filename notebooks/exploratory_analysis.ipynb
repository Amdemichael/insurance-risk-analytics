```python
# Import required libraries and the eda module
import sys
import os
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..', 'src')))  # Adjust path to src
import eda
import pandas as pd
import matplotlib.pyplot as plt

# Enable inline plotting in notebook
%matplotlib inline

# Load data with absolute path
data_path = os.path.abspath(os.path.join(os.getcwd(), '..', 'data', 'MachineLearningRating_v3.txt'))
df = eda.load_data(data_path)
```
